Muestra un gr√°fico de barras con el total de productos por categoria

```python
import pandas as pd
import matplotlib.pyplot as plt

datos=pd.read_csv('products.csv')
df=datos[['unitsInStock','categoryID']]
df.head(25).plot.bar(x="unitsInStock", y="categoryID")
plt.show()
```

Muestra un gr√°fico circular con el precio medio por categoria 

```python
import pandas as pd
import matplotlib.pyplot as plt

def consumirCircular():
    datos=pd.read_csv('products.csv')
    df=datos[['unitsInStock','categoryID']]
    df.head(25).plot.pie(x="unitsInStock", y="categoryID")
    plt.show()

consumirCircular()
```

Muestra un histograma con los precios de los productos

```python
import pandas as pd
import matplotlib.pyplot as plt

def  histograma():
    datos=pd.read_csv('products.csv')
    df=datos[['unitPrice','categoryID']]
    df.head(7).unitPrice.plot.hist()
    print(df.sample(5))
histograma()
```
